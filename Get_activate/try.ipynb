### 看一下提取出的激活and合并保存测试/验证激活

```python
from datasets import Dataset
import pandas as pd
import torch as th
import numpy as np
```

```python
# 读取 .arrow 文件
dataset = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-test-data1.arrow')
dataset2 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-test-data2.arrow')
dataset3 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-test-data3.arrow')
dataset4 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-test-data4.arrow')
dataset8 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-test-data5.arrow')
dataset9 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-test-data6.arrow')
dataset5 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-valid-data1.arrow')
dataset6 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-valid-data2.arrow')
dataset7 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-valid-data3.arrow')
dataset10 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-valid-data4.arrow')
dataset11 = Dataset.from_file('dataset/esm2_ctrlprot_dataset/bp-valid-data5.arrow')
# 将数据集转换为 Pandas DataFrame
df = dataset.to_pandas()
# 追加dataset2到df
df = pd.concat([df, dataset2.to_pandas()], axis=0)
df_test = pd.concat([df, dataset3.to_pandas()], axis=0)
df_test = pd.concat([df_test, dataset4.to_pandas()], axis=0)
df_test = pd.concat([df_test, dataset8.to_pandas()], axis=0)
df_test = pd.concat([df_test, dataset9.to_pandas()], axis=0)


df_valid = dataset5.to_pandas()
df_valid = pd.concat([df_valid, dataset6.to_pandas()], axis=0)
df_valid = pd.concat([df_valid, dataset7.to_pandas()], axis=0)
df_valid = pd.concat([df_valid, dataset10.to_pandas()], axis=0)
df_valid = pd.concat([df_valid, dataset11.to_pandas()], axis=0)

print(df_valid.shape)
print(df_test.shape)
df_valid.head(3)
```

```python
df_valid = df_valid[['entry', 'head_activation', 'mlp_activation']]
df_test = df_test[['entry', 'head_activation', 'mlp_activation']]
# df的entry列更名为proteins
df_valid = df_valid.rename(columns={'entry': 'proteins'})
df_test = df_test.rename(columns={'entry': 'proteins'})
```

```python
df_test.head(3)
```

```python
origin_test_df = pd.read_pickle('../deepgo2/data/bp/test_data.pkl')
origin_valid_df = pd.read_pickle('../deepgo2/data/bp/valid_data.pkl')
# 统计origin_df的proteins列中的字符串是否有重名
# print(origin_df['proteins'].duplicated().sum())
# # 输出origin_df的proteins列中名为FPGS1_ARATH，ASPH_HUMAN，XANP_ECOLI的行
# print(origin_df[origin_df['proteins'].isin(['FPGS1_ARATH', 'ASPH_HUMAN', 'XANP_ECOLI'])])

# 去除origin_test_df和origin_valid_df的interpros和esm列
origin_test_df = origin_test_df.drop(columns=['interpros', 'esm'])
origin_valid_df = origin_valid_df.drop(columns=['interpros', 'esm'])
print(origin_test_df.shape)
print(origin_valid_df.shape)
origin_test_df.head(3)
```

```python
# 将df和origin_df合并，根据proteins列合并
df_test = pd.merge(df_test, origin_test_df, on='proteins')
print(df_test.shape)
df_valid = pd.merge(df_valid, origin_valid_df, on='proteins')
print(df_valid.shape)
df_valid.head(3)
```

```python
df_valid.to_pickle('../deepgo2/data/bp/valid_esm_data.pkl')
df_test.to_pickle('../deepgo2/data/bp/test_esm_data.pkl')
```

### 合并保存训练数据激活

```python
dataset_train = Dataset.from_file('dataset/esm2_ctrlprot_dataset/data-00000-of-00089.arrow')
df_train = dataset_train.to_pandas()
# 用for循环读取剩余的35个.arrow文件，并将其转换为Pandas DataFrame，然后追加到df_train
for i in range(1, 89):
    dataset = Dataset.from_file(f'dataset/esm2_ctrlprot_dataset/data-000{i:02}-of-00089.arrow')
    df_train = pd.concat([df_train, dataset.to_pandas()], axis=0)

print(df_train.shape)
df_train.head(3)
```

```python
for i, row in enumerate(df_train.itertuples()):
    print(row.head_activation.shape)
    print(row.head_activation[31])
    if i==3:
        break
for i, row in enumerate(df_train.itertuples()):
    print(row.mlp_activation.shape)
    print(row.mlp_activation[31])
    if i==3:
        break
```

```python
df_train = df_train[['entry','sequence', 'head_activation','mlp_activation']]
origin_train_df = pd.read_pickle('../deepgo2/data/bp/train_data.pkl')
origin_train_df = origin_train_df.drop(columns=['interpros', 'esm'])
print(origin_train_df.shape)
# 统计origin_df的proteins列中的字符串是否有重名
print(origin_train_df['proteins'].duplicated().sum())
origin_train_df.head(3)
```

```python
df_train = df_train[['entry', 'head_activation','mlp_activation']]
df_train = df_train.rename(columns={'entry': 'proteins'})
df_train = pd.merge(df_train, origin_train_df, on='proteins')
print(df_train.shape)
df_train.head(3)
```

```python
df_train.to_pickle('../deepgo2/data/bp/train_esm_data.pkl')
```
