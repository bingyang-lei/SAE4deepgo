```python
from datasets import Dataset
import pandas as pd
import torch as th
import numpy as np
```

```python
# 读取 .arrow 文件
dataset = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-test-activation1.arrow')
dataset2 = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-test-activation2.arrow')
dataset3 = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-test-activation3.arrow')
dataset4 = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-test-activation4.arrow')
dataset5 = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-valid-activation1.arrow')
dataset6 = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-valid-activation2.arrow')
dataset7 = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/cc-valid-activation3.arrow')
# 将数据集转换为 Pandas DataFrame
df = dataset.to_pandas()
# 追加dataset2到df
df = pd.concat([df, dataset2.to_pandas()], axis=0)
df_test = pd.concat([df, dataset3.to_pandas()], axis=0)
df_test = pd.concat([df_test, dataset4.to_pandas()], axis=0)

df_valid = dataset5.to_pandas()
df_valid = pd.concat([df_valid, dataset6.to_pandas()], axis=0)
df_valid = pd.concat([df_valid, dataset7.to_pandas()], axis=0)
print(df_valid.shape)
print(df_test.shape)
df_valid = df_valid[['entry', 'mlp_activation']]
df_test = df_test[['entry', 'mlp_activation']]
# df的entry列更名为proteins
df_valid = df_valid.rename(columns={'entry': 'proteins'})
df_test = df_test.rename(columns={'entry': 'proteins'})
```

```python
origin_test_df = pd.read_pickle('../deepgo2/data/cc/test_data.pkl')
origin_valid_df = pd.read_pickle('../deepgo2/data/cc/valid_data.pkl')
# 统计origin_df的proteins列中的字符串是否有重名
# print(origin_df['proteins'].duplicated().sum())
# # 输出origin_df的proteins列中名为FPGS1_ARATH，ASPH_HUMAN，XANP_ECOLI的行
# print(origin_df[origin_df['proteins'].isin(['FPGS1_ARATH', 'ASPH_HUMAN', 'XANP_ECOLI'])])

# 去除origin_test_df和origin_valid_df的interpros和esm列
origin_test_df = origin_test_df.drop(columns=['interpros', 'esm'])
origin_valid_df = origin_valid_df.drop(columns=['interpros', 'esm'])
print(origin_test_df.shape)
print(origin_valid_df.shape)
origin_test_df.head(3)
```

```python
# 将df和origin_df合并，根据proteins列合并
df_test = pd.merge(df_test, origin_test_df, on='proteins')
print(df_test.shape)
df_valid = pd.merge(df_valid, origin_valid_df, on='proteins')
print(df_valid.shape)
df_valid.head()
```

```python
dataset_train = Dataset.from_file('dataset/protgpt2_ctrlprot_dataset/data-00000-of-00049.arrow')
df_train = dataset_train.to_pandas()
# 用for循环读取剩余的35个.arrow文件，并将其转换为Pandas DataFrame，然后追加到df_train
for i in range(1, 49):
    dataset = Dataset.from_file(f'dataset/protgpt2_ctrlprot_dataset/data-000{i:02}-of-00049.arrow')
    df_train = pd.concat([df_train, dataset.to_pandas()], axis=0)

print(df_train.shape)
df_train.head(3)
```

```python
df_train = df_train[['entry','sequence', 'mlp_activation']]
origin_train_df = pd.read_pickle('../deepgo2/data/cc/train_data.pkl')
origin_train_df = origin_train_df.drop(columns=['interpros', 'esm'])
print(origin_train_df.shape)
# 统计origin_df的proteins列中的字符串是否有重名
print(origin_train_df['proteins'].duplicated().sum())
origin_train_df.head(3)
```

```python
df_train = df_train[['entry', 'mlp_activation']]
df_train = df_train.rename(columns={'entry': 'proteins'})
df_train = pd.merge(df_train, origin_train_df, on='proteins')
print(df_train.shape)
df_train.head(3)
```
